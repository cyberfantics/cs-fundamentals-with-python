    <p style="color:#e5e7eb; margin:0;">
        <strong>016.</strong> Ask the user if it is raining.
        Convert their answer to lower case so the input is case-insensitive.
        If they answer <code style="color:#38bdf8;">yes</code>, ask if it is windy.
        If they answer <code style="color:#38bdf8;">yes</code> again, display
        <code style="color:#38bdf8;">It is too windy for an umbrella</code>.
        Otherwise, display
        <code style="color:#38bdf8;">Take an umbrella</code>.
        If the user did not answer yes to the first question,
        display
        <code style="color:#38bdf8;">Enjoy your day</code>.
    </p>
</div>


```python
rain = input("Is it is raining ").lower().strip()
if rain == 'yes' or rain == 'y' or rain == 'true':
    windey = input("Is it is windy ").lower().strip()
    if windey == 'yes' or windey == 'y' or windey == 'true':
        print("It is too windy for an umbrella")
    else:
        print("Take an Umbrella")
else:
    print("Enjoy your day")
```
