## 006: Ask how many slices of pizza the user started with and ask how many slices they have eaten. Work out how many slices they have left and display the answer in a userfriendly format. 

```python
start = int(input("How many slices of pizza you have started "))
eaten = int(input("How many slices of pizza you have eaten "))
print(f"You initially have {start} slices and you eat {eaten} of them. \n So now you only have {start-eaten} slices")
```

## 009 Write a program that will ask for a number of days and then will show how many hours, minutes and seconds are in that number of days. 

```python
days = int(input("Enter number of days "))
hours = days * 24
minutes = hours * 60
seconds = minutes * 60
print(f'Their are {hours} hours in {days} days.\n Their are {minutes} minutes in {hours} hours\n And their are {seconds} seconds in {minutes} minutes')
```

## 010  There are 2.204 pounds in a kilogram. Ask the user to enter a weight in kilograms and convert it to pounds. 

```python
ponds = 2.204
weight = int(input("Enter weight "))
print(f'Their are {ponds * weight} ponds in {weight} kg')
```

## 011  Task the user to enter a number over 100 and then enter a number under 10 and tell them how many times the smaller number goes into the larger number in a user-friendly format.

```python
over = int(input("Enter number over 10 "))
under = int(input("Enter number under 10 "))
answer = over // under
print(f'Smaller {under} goes to larger {over} over {answer} times')
```
