    <p style="color:#e5e7eb; margin:0;">
        <strong>026.</strong> Pig Latin rules:
        If a word starts with a consonant, move the first letter to the end
        and add <strong>“ay”</strong>.
        If the word starts with a vowel, add <strong>“way”</strong> to the end.
        Ask the user to enter a word and convert it to Pig Latin.
        Display the result in lowercase.
    </p>
</div>


```python
text = input("Enter your sentence ")
words = text.split(' ')
newT = ''
for word in words:
    if word[0].lower() in ['a', 'e', 'i', 'o', 'u']:
        newT += word + 'ay '
    else:
        newT += word + 'way '
print(newT)
```

    <p style="color:#e5e7eb; margin:0;">
        <strong>059.</strong> Display five colours and ask the user to pick one. 
        If they pick the correct colour, display “Well done”. 
        If incorrect, give a witty clue involving the correct colour (e.g., “I bet you are GREEN with envy”) and ask again. 
        Repeat until the user guesses correctly.
    </p>
</div>


```python
colours = ["RED", "BLUE", "GREEN", "YELLOW", "PURPLE"]
correct_colour = random.choice(colours)

while True:
    print("Pick a colour from the list below:")
    for colour in colours:
        print("-", colour)

    guess = input("Your choice: ").strip().upper()

    if guess == correct_colour:
        print("Well done!")
        break
    else:
        clues = {
            "RED": "I can see you are feeling a little RED-faced!",
            "BLUE": "Don’t feel BLUE, try again!",
            "GREEN": "I bet you are GREEN with envy!",
            "YELLOW": "Are you feeling a bit YELLOW? Try again!",
            "PURPLE": "That guess wasn’t very PURPLE-icious!"
        }
        print(clues[correct_colour])
```
